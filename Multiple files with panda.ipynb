##Olympic sports

```python
file="dictionary.csv"
countries_df= pd.read_csv(file, decimal=",")
countries_df
```

```python
file="summer.csv"
summer_df= pd.read_csv(file, decimal=",")
summer_df
```

```python
summer_df_2=summer_df.rename(columns={'Country':'Code'})
```

```python
summer_df
```

```python
file="winter.csv"
winter_df= pd.read_csv(file, decimal=",")
winter_df
```

```python
winter_df.rename(columns={'Country':'Code'},inplace='True')
winter_df
```

```python
summer_countries_df=summer_df_2.merge(countries_df,on='Code',how='left')
```

```python
winter_countries_df=winter_df.merge(countries_df,on='Code',how='left')
```

```python
summer_countries_df['Season']='Summer'
summer_countries_df
```

```python
winter_countries_df['Season']='Winter'
winter_countries_df
```

```python
all_df=pd.concat([winter_countries_df,summer_countries_df])
```

```python
all_df
```

```python
all_df_1984=all_df.loc[all_df['Year']>=1984]
all_df_1984
```

```python
all_group=all_df_1984.groupby(['Code']).count()[['Medal']]
plot=all_group.sort_values(by='Medal',ascending=False).head(10)
```

```python
plot.plot()
```
